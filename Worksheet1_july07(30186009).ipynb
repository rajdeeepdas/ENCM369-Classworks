```python
#2a Boolean equation: F = AB + CA'
#2b&c
def lock_logic(A, B, C):
    output = (A and B) or (C and not A)
    return int(output)

print(lock_logic(0, 1, 1))

"""
By hand for A = 0, B = 1, C = 1:
F = AB + CA'
F = (0)(1) + (1)(1)
F = 0 + 1
F = 1
Therefore, the lock opens because the output is 1.

"""
```

```python
"""
Signed two’s complement interpretation: Since the most significant bit is 1, the number is negative.
Original: 1111 0110
Invert:   0000 1001
Add 1:    0000 1010 = 10
Signed value = -10
"""
result = int('11110110', 2)
# Convert result back to signed value
if result & 0x80:                # Check if sign bit (bit 7) is 1 → means negative
    signed_result = result - 256 # Convert to negative value by subtracting 256
else:
    signed_result = result       # Otherwise it's already positive
print(signed_result)
```
